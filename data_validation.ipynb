```python
import csv
from graphviz import Digraph
from IPython.display import display

# Main nodes
wastewater = {"name":"Wastewater","size":1}
cases = {"name":"Cases","size":1}
niwa = {"name":"NIWA","size":1}


#Create empty lists for sub-nodes
weather_factors = []
cases_factors = []
wastewater_factors = []


# define a funtion to read factors from the data file
def read_factors_from_csv_list(file_paths, factor_size=0.5):
    factors = []
    for file_path in file_paths:
        with open(file_path, mode='r') as csvfile:
            csv_reader = csv.reader(csvfile)
            headers = next(csv_reader)
            for header in headers:
                factors.append({"name": header, "size": factor_size})
    return factors

niwa_files = ['dataset/modelling_data/niwa/weather_data.csv']
case_files = ['dataset/modelling_data/cases/covid-case-counts.csv','dataset/modelling_data/cases/weekly-deaths.csv', 'dataset/modelling_data/cases/weekly-hospitalisations-for-covid.csv']
wastewater_files = ['dataset/modelling_data/wastewater/daily_ww_test_result.csv']

weather_factors = read_factors_from_csv_list(niwa_files)
cases_factors = read_factors_from_csv_list(case_files)
wastewater_factors = read_factors_from_csv_list(wastewater_files)

# Create the graph
graph = Digraph(engine="dot")
graph.attr(rankdir="TB", nodesep="0.5", ranksep="1.5")

# Add the main nodes
graph.node(wastewater["name"], width=str(wastewater["size"]), height=str(wastewater["size"]))
graph.node(cases["name"], width=str(cases["size"]), height=str(cases["size"]))
graph.node(niwa["name"], width=str(niwa["size"]), height=str(niwa["size"]))
# Add the sub-nodes and edges
with graph.subgraph(name="cluster_niwa") as niwa_cluster:
    niwa_cluster.attr(label="NIWA", color="gray90")
    for factor in weather_factors:
        niwa_cluster.node(factor["name"], width=str(factor["size"]), height=str(factor["size"]))
        graph.edge(factor["name"], niwa["name"], dir="none", weight="100")
with graph.subgraph(name="cluster_cases") as cases_cluster:
    cases_cluster.attr(label="Cases", color="gray90")
    for factor in cases_factors:
        cases_cluster.node(factor["name"], width=str(factor["size"]), height=str(factor["size"]))
        graph.edge(factor["name"], cases["name"], dir="none", weight="100")
with graph.subgraph(name="cluster_wastewater") as wastewater_cluster:
    wastewater_cluster.attr(label="Wastewater", color="gray90")
    for factor in wastewater_factors:
        wastewater_cluster.node(factor["name"], width=str(factor["size"]), height=str(factor["size"]))
        graph.edge(factor["name"], wastewater["name"], dir="none", weight="100")
# Display the graph
display(graph)
```

```python
import os
import pandas as pd
import pandas.errors as pd_errors
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import timedelta
import dateutil.parser
import re
def collect_files(path):
    data_files = []
    for root, _, files in os.walk(path):
        for file in files:
            if file.endswith('.csv'): # or other file format, if needed
                data_files.append(os.path.join(root, file))
    return data_files 
def is_date(string, fuzzy=True):
    if not isinstance(string, str):
        return False
    try: 
        dateutil.parser.parse(string, fuzzy=fuzzy)
        return True
    except ValueError:
        return False
def is_date_column(column, min_proportion=0.5, max_rows_to_check=5):
    column_name = column.name.lower()
    if column_name == 'period' or column_name == 'date':
        return True
    date_values = sum(is_date(value) for value in column[:max_rows_to_check])
    return date_values / max_rows_to_check >= min_proportion
def process_files(files):
    feature_dict = {}

    for file in files:
        try:
            df = pd.read_csv(file, low_memory=False)
            df.columns = [col.lower() for col in df.columns]
            date_column = None
            if 'date' in df.columns:
                date_column = 'date'
            elif 'period' in df.columns:
                date_column = 'period'
            else:
                for col in df.columns:
                    if is_date_column(df[col]):
                        date_column = col
                        break
            if date_column:
                df[date_column] = pd.to_datetime(df[date_column],errors ='coerce').dropna()
                for feature in df.columns:
                    if feature != date_column :
                        if feature in feature_dict:
                            min_date = min(df[date_column].min(), feature_dict[feature][0])
                            max_date = max(df[date_column].max(), feature_dict[feature][1])
                            feature_dict[feature] = (min_date, max_date)
                        else: feature_dict[feature] = (df[date_column].min(), df[date_column].max())
        except pd_errors.EmptyDataError:
            print(f"Skipping empty file: {file}")
    return feature_dict
def plot_feature_ranges(feature_dict):
    fig, ax = plt.subplots(figsize=(18, 12))
    y_labels = []
    y_ticks = []
    for i, (feature, date_range) in enumerate(feature_dict.items()):
        
            ax.barh(i, date_range[1] - date_range[0], left=date_range[0], height=0.8)
            y_labels.append(feature)
            y_ticks.append(i)

    ax.set_yticks(y_ticks)
    ax.set_yticklabels(y_labels)
    ax.set_xlabel('Date')
    ax.set_ylabel('Features')
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    ax.xaxis.set_major_locator(mdates.YearLocator())
    custom_min_date = pd.to_datetime('2006-06-30')
    custom_max_date = pd.to_datetime('2023-12-31')
    ax.set_xlim(custom_min_date, custom_max_date)
    plt.tight_layout()
    plt.savefig('time_range.png')
    plt.show()
    
data_files = collect_files('dataset/modelling_data')
feature_dict = process_files(data_files)
plot_feature_ranges(feature_dict)
```

```python
import os
import pandas as pd
import pandas.errors as pd_errors
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import timedelta
import dateutil.parser
import re
def collect_files(path):
    data_files = []
    for root, _, files in os.walk(path):
        for file in files:
            if file.endswith('.csv'): # or other file format, if needed
                data_files.append(os.path.join(root, file))
    return data_files 
def is_date(string, fuzzy=True):
    if not isinstance(string, str):
        return False
    try: 
        dateutil.parser.parse(string, fuzzy=fuzzy)
        return True
    except ValueError:
        return False
def is_date_column(column, min_proportion=0.5, max_rows_to_check=10):
    column_name = column.name.lower()
    if column_name == 'period' or column_name == 'date':
        return True
    date_values = sum(is_date(value) for value in column[:max_rows_to_check])
    return date_values / max_rows_to_check >= min_proportion


def process_files(files):
    combined_df = None

    for file in files:
        try:
            if ".ipynb_checkpoints" in file:
                continue
            
            df = pd.read_csv(file, low_memory=False)

            df.columns = [col.lower() for col in df.columns]
            date_column = None
            if 'date' in df.columns:
                date_column = 'date'
            elif 'period' in df.columns:
                date_column = 'period'
            
            else:
                for col in df.columns:
                    if is_date_column(df[col]):
                        date_column = col
                        break
            if date_column:
                
                df[date_column] = pd.to_datetime(df[date_column], errors='coerce')
                print(f"Processing file: {file}, date column: {date_column}")
                df.dropna(subset=[date_column],inplace=True)
                df.set_index(date_column, inplace=True)
                df = df.resample('D').mean()
                df.interpolate(method='linear', limit_area='inside', inplace=True)
                if combined_df is None:
                    combined_df = df
                else:
                    combined_df = pd.concat([combined_df, df], axis=1, join='outer')
        except pd.errors.EmptyDataError:
            print(f"Skipping empty file: {file}") 
    # Remove duplicate columns (if any)
    combined_df = combined_df.loc[:, ~combined_df.columns.duplicated()]
    combined_df.dropna(axis=1, how='all', inplace=True)
    return combined_df
def plot_feature_time_series(df): 
    plt.figure(figsize=(100,80))
    ax=df.plot(subplots=True, layout=(-1, 1), sharex=True, figsize=(100,80))
    min_date = df.index.min()
    max_date=df.index.max()
    for axes in ax:
        for a in axes:
            a.set_xlim(min_date,max_date)
            label = a.get_ylabel()
            if label in df.columns:
                min_value =df[label].min()
                max_value =df[label].max()
                a.set_ylim(min_value, max_value)
    plt.savefig('time_series.png')
    plt.show()
    
    
data_files = collect_files('dataset/modelling_data')
combined_df = process_files(data_files)
plot_feature_time_series(combined_df)



```
